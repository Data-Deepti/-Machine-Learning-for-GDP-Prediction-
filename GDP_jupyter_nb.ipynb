**importing libraries : numpy , pandas ,sci-kit learn**

```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

**loading GDP data (source: OGD Platform India)**

```python
df = pd.read_csv (r'C:\Users\Deepti Anand\Desktop\GDP_.csv' , encoding='unicode_escape')
```

**EDA(Exploratory Data Analysis)**

**Preparing Data**

```python
x = df[['Agriculture_forestry_fishing'  ,'mining' ,'manufacturing' ,'electricity' ,'construction' ,'trade_services','financial_realestate' ,'publicadministration_defence' ,'GVA_basic_price','GFCF' ,'PFCE','GFCF' ,'CIS','valuables','Exports' ,'Imports', 'discrepencies']]
```

```python
x
```

```python
y=df['GDP__']
y
```

**Splitting Data**

```python
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.2 ,random_state= 40)
```

```python
x_train
```

```python
x_test
```

**Initialize and Fit the model**

```python
model=LinearRegression()
model.fit(x_train , y_train)
```

**Evaluating the model and checking for it's Accuracy**

```python
y_pred=model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
mse = metrics.mean_squared_error(y_test, y_pred)
r_squared = metrics.r2_score(y_test, y_pred)
```

```python
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r_squared}')
```
